```python
%matplotlib inline

import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv("hema_data.txt" ,  sep = "\t", header = 0, index_col=0)
df.head()
```

```python
def plot_heatmap(df, title):
    plt.pcolor(df)
    plt.title("Sorted")
    plt.xlabel("Cell types")
    plt.ylabel("Gene")
    plt.xticks(np.arange(0.5, df.shape[1], 1), df.columns);
```

```python
from scipy.cluster.hierarchy import linkage,leaves_list,dendrogram

Z =linkage(df)
idx = leaves_list(Z)

df = df.iloc[idx]

df.head()
```

```python
plot_heatmap(df, 'Sorted')
```

```python
x = df["poly"]
y = df["CFU"]
ngenes = df.shape[0]
```

```python
plt.scatter(x,y);
```

```python
import random
random.choice
#v = [i*3 for i in range(5)] # list comprehension

c = [random.choice([0,1]) for i in range(ngenes)]
plt.scatter(x,y, c=c, cmap='Paired', vmin=0,vmax=12)
plt.colorbar()
```

```python
roi = y > x
c = np.zeros([ngenes]) + 6
c[roi] = 9
plt.scatter(x,y, c=c, cmap='Paired', vmin=0,vmax=12)
```

```python
plt.scatter(x,y, c=df['unk'], cmap = 'summer');
plt.colorbar()
```

```python
plt.figure(figsize=[12,4])
plt.subplot(1,3,1)
plot_heatmap(df,'Sorted')
plt.subplot(1,3,2)
roi = y > x
c = np.zeros([ngenes]) + 6
c[roi] = 9
plt.scatter(x,y, c=c, cmap='Paired', vmin=0,vmax=12)
plt.subplot(1,3,3)
plt.scatter(x,y, c=df['unk'], cmap = 'summer');
plt.colorbar()
plt.tight_layout();
plt.savefig("summary_fig.pdf")
```
